```python
import random
my_array = ['A', 'A#/Bb', 'B', 'C', 'C#/Db', 'D', 'D#/Eb','E','F',
            'F#/Gb','G','G#/Ab']
a =  ''
b =  ''
randomTones = ''
toneSplit = [None] * 2
```

```python
def getRandom():
    # Generate a random int number within a specific range
    random_number1 = random.randint(0, len(my_array)-1)
    random_number2 = random.randint(0, len(my_array)-1)
    if random_number1 == random_number2:
        random_number2 = random.randint(0, len(my_array)-1)
    else:
        return(f'{str(my_array[random_number1])},{str(my_array[random_number2])}')
    
    
```

```python
def getNumOfSemiTone(start_value,end_value):
    found_start = False
    found_end = False
    start_index = 0
    end_index = 0

    for i, value in enumerate(my_array):
        if value == start_value:
            start_index = i
            found_start = True
        elif value == end_value:
            end_index = i
            found_end = True
        elif found_start and found_end:
            break

    if found_start and found_end:
        if start_index <= end_index:
            index = end_index - start_index
        else:
            index = len(my_array) - (start_index - end_index)
        if index > len(my_array) / 2:##traverse search array
            index = len(my_array) - index
         
        return index
```
